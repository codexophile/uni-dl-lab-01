2.  Create a random 100000x1 array from the same exponential distribution and visualize the
distribution histogram along with uniform and normal distributions. Use the code given
below and do the necessary changes. Change the number of bins and the plot view to make
the visualization better.

```python
import numpy as np
import matplotlib.pyplot as plt

exp_dist = np.random.exponential(1.0, 100000)
uniform_dist = np.random.rand(100000)
normal_dist = np.random.randn(100000)

plt.hist(exp_dist, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniform_dist, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_dist, density=True, bins=100, histtype="step", color="green", label="Normal")

plt.axis([-4, 6, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

3. Using matplotlib, plot the Z = X^2 + Y^2 in a 3D plot. Limit the x and y values to [-5, 5].

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

fig = plt.figure()
ax = plt.axes(projection='3d')

x_vals = np.linspace(-5, 5, 50)
y_vals = np.linspace(-5, 5, 50)

X, Y = np.meshgrid(x_vals, y_vals)

Z = X**2 + Y**2

ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title("Z = X**2 + Y**2")
plt.show()
```
